```python
# first create models
# for some (kn and grb) I used a theoretical model
# for sn1a and TDE I used real transients from ZTF


# KN model

def model_kn(tmax = 370):
    t_original = np.arange(0.01, tmax, .1)
    kn_model = SimpleKilonovaLightCurveModel(sample_times = t_original)

    # run the simulation
    bestfit_params = {
        "luminosity_distance": 46.4894661695928,
        "beta": 3.6941470839046575,
        "log10_kappa_r": 1.6646575950692677,
        "KNtimeshift": 0.183516607107672,
        "log10_vej": -0.431500234170688,
        "log10_Mej": -1.4122890,
    }
    _, knmag = kn_model.generate_lightcurve(t_original, bestfit_params)

    # using linear interpolation
    kn_r = interp1d(t_original,knmag['r'])
    kn_g = interp1d(t_original,knmag['g'])
    return kn_g,kn_r


# GRB model
def model_grb(tmax = 370):
# Jet Parameters
    Z = {'jetType':     grb.jet.Gaussian,     # Gaussian jet
         'specType':    0,                  # Basic Synchrotron Emission Spectrum
         'thetaObs':    0.1,   # Viewing angle in radians
         'E0':          1.0e53, # Isotropic-equivalent energy in erg
         'thetaCore':   0.05,    # Half-opening angle in radians
         'thetaWing':   0.4,    # Outer truncation angle
         'n0':          1.0e-3,    # circumburst density in cm^{-3}
         'p':           2.2,    # electron energy distribution index
         'epsilon_e':   0.1,    # epsilon_e
         'epsilon_B':   0.0001,   # epsilon_B
         'xi_N':        1.0,    # Fraction of electrons accelerated
         'd_L':         1.36e26, # Luminosity distance in cm
         'z':           0.5}   # redshift


    # Time and Frequencies

    t_original = np.arange(0.1, tmax, .1)

    nuOr = 2.99e8/625e-9
    nuOg = 2.99e8/480e-9

    # Calculate!

    FnuOr = grb.fluxDensity(t_original * grb.day2sec, nuOr, **Z)
    FnuOg = grb.fluxDensity(t_original * grb.day2sec, nuOg, **Z)

    # using linear interpolation
    grb_r = interp1d(t_original,-2.5*np.log10(FnuOr/1e3)+8.9)
    grb_g = interp1d(t_original,-2.5*np.log10(FnuOg/1e3)+8.9)

    return grb_g,grb_r


# TDE model

def model_tde(tmax = 370, tde_file = './TDE_example.csv'):
    tde_lc = pd.read_csv(tde_file)  
    tde_lc['JD'] = tde_lc['JD'] - np.min(tde_lc['JD'])
    
    # fixing first and last data point
    df_n = tde_lc[-1:].copy()
    df_n['fid'] = 1
    tde_lc = tde_lc.append(df_n,ignore_index=True)

    df_n = tde_lc[-1:].copy()
    df_n['JD'] = 380
    df_n['magpsf'] = 21.2
    df_n['fid'] = 2
    tde_lc = tde_lc.append(df_n,ignore_index=True)

    tde_lc = tde_lc.sort_values(by=['JD'])
#     plt.plot(tde_lc['JD'][tde_lc['fid']==1],tde_lc['magpsf'][tde_lc['fid']==1],color='g')
#     plt.plot(tde_lc['JD'][tde_lc['fid']==2],tde_lc['magpsf'][tde_lc['fid']==2],color='r')
#     plt.ylim(21.5,18)

    # # using linear interpolation
    TDE_g = interp1d(tde_lc['JD'][tde_lc['fid']==1],tde_lc['magpsf'][tde_lc['fid']==1])
    TDE_r = interp1d(tde_lc['JD'][tde_lc['fid']==2],tde_lc['magpsf'][tde_lc['fid']==2])

    return TDE_g, TDE_r

# SN1a

def model_sn1a(tmax = 370,sn1a_file = './SN1a_example.csv'):
    
    sn1a_lc = pd.read_csv(sn1a_file)  
    # sn1a_lc
    sn1a_lc = sn1a_lc[sn1a_lc['filter'].isin(['ztfg','sdssg','ztfr','sdssr',])]
    sn1a_lc.dropna(subset = ["mag"], inplace=True)
    sn1a_lc['mjd'] = sn1a_lc['mjd'] - np.min(sn1a_lc['mjd'])
    sn1a_lc = sn1a_lc.sort_values(by=['mjd'])
    
    # first det

    df_n = sn1a_lc[:1].copy()
    if 'r' in df_n['filter'].values[0]:
        df_n['filter'] ='sdssg'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)
    elif 'g' in df_n['filter'].values[0]:
        df_n['filter'] ='sdssr'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)
    else:
        print('here',df_n['filter'])
        df_n['filter'] ='sdssr'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)
        df_n['filter'] ='sdssg'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)

    sn1a_lc = sn1a_lc.sort_values(by=['mjd'])

    # # # last detection
    df_n = sn1a_lc[-1:].copy()
    if 'r' in df_n['filter'].values[0]:
        df_n['filter'] ='sdssg'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)
    elif 'g' in df_n['filter'].values[0]:
        df_n['filter'] ='sdssr'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)
    else:
        print('here',df_n['filter'])
        df_n['filter'] ='sdssr'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)
        df_n['filter'] ='sdssg'
        sn1a_lc = sn1a_lc.append(df_n,ignore_index=True)

    sn1a_lc = sn1a_lc.sort_values(by=['mjd'])

    g_flit = (sn1a_lc['filter']=='sdssg') | (sn1a_lc['filter']=='ztfg')
    r_flit = (sn1a_lc['filter']=='sdssr') | (sn1a_lc['filter']=='ztfr')
    
    # # # using linear interpolation
    SN1a_g = interp1d(sn1a_lc['mjd'][g_flit],sn1a_lc['mag'][g_flit])
    SN1a_r = interp1d(sn1a_lc['mjd'][r_flit],sn1a_lc['mag'][r_flit])

    return SN1a_g, SN1a_r
```

```python
# Creating an alert


def create_arvo(lc,ii,
                parsed_schema= parse_schema(json.loads(open("ztf_alert.avsc").read())),
                alert_template='./1105522281015015000.avro'):
    '''
    lc: list that contains the following arrays characterizing the light curve jd,mag,filter,maglim
    ii: int that numbers the output alert as mztf_fastavro.test{ii}.avro
    parsed_schema: ZTF avro schema'''
    
    temp_dt,temp_mag,filt_temp,temp_maglim,transient_type = lc
    # model/lightcurve features
    mod_mag = temp_mag 
    mod_days = temp_dt
    temp_detections = np.where((mod_mag < temp_maglim) == True)[0]
    
    idx_jdstart = temp_detections[0]
    idx_jdend = temp_detections[-1]
    
    # ra dec + PS1
    ra,dec = random.uniform(0,359),random.uniform(-10,70)
    coord = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')
    c = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')
    
    tablePS1 = False
    if tablePS1:
        tab = ps1_sample(ra=ra, dec=dec)
        if len(tab)>0:
            separation = (c.separation(SkyCoord(ra=tab['raMean']*u.degree, dec=tab['decMean']*u.degree, frame='icrs')).to(u.arcsec)).value
            tab['distpsnr'] = separation
            tab = tab[np.argsort(tab['distpsnr'])]
            tab['sgscore'] = np.abs(tab['iMeanKronMag']-tab['iMeanPSFMag']) 

            keys = ['sgmag', 'srmag', 'simag', 'szmag', 'sgscore', 'distpsnr']
            key_tab = ['gMeanPSFMag', 'rMeanPSFMag', 'iMeanPSFMag', 'zMeanPSFMag', 'sgscore', 'distpsnr']
            sg = tab['sgscore']
            # need to check the sgscore
            if len(sg[sg>0.2]):
                sg[sg>0.2] = sg[sg>0.2] / np.max(sg[sg>0.2])
            tab['sgscore'] = np.abs(1.-np.round(sg,4))
    # need to fix PS1 query
        else:
            tablePS1 = False
    # modify candidate
    with open(alert_template, 'rb') as fo:
        avro_reader = reader(fo)
        for record in avro_reader:
            record['candidate']['class'] = transient_type
            record['candidate']['jd'] = mod_days[idx_jdend]
            record['candidate']['jdstarthist'] = mod_days[idx_jdstart]
            record['candidate']['fid'] = filt_temp[idx_jdend] # change filter
            record['candidate']['diffmaglim'] = temp_maglim[idx_jdend]
            record['candidate']['magpsf'] = mod_mag[idx_jdend]
            record['candidate']['magap']  = mod_mag[idx_jdend]
            record['candidate']['b']  = c.galactic.b.to(u.degree).value
            record['candidate']['drbscore'] = random.uniform(0,0.3)
            record['candidate']['nbad'] = 3
            record['candidate']['fwhm'] = random.uniform(0.5,5)
            # need to query MPC
            record['candidate']['ssdistnr'] = -999
            record['candidate']['ssmagnr'] = -999
            
            # need to get the sgscore from ml pipeline
            if tablePS1:
                for i in range(1,4):
                    for k,kt in np.array([keys,key_tab]).T:
                        record['candidate'][k+str(i)] = tab[kt] 
#             else:
#                 for i in range(1,4):
#                     for k,kt in np.array([keys,key_tab]).T:
#                         record['candidate'][k+str(i)] = -99
            prv_collection = []

            for i in range(idx_jdend):
                temp_prv = record['prv_candidates'][0].copy()
                temp_prv['jd'] = temp_dt[i]
                temp_prv['fid'] = filt_temp[i]
                temp_prv['diffmaglim'] = temp_maglim[i]
                if mod_mag[i]>temp_maglim[i]:
                    temp_prv['magpsf'] = -99
                else:
                    temp_prv['magpsf'] = mod_mag[i]
                prv_collection.append(temp_prv)    

            record['prv_candidates'] = prv_collection

    with open('mztf_fastavro.test%d.avro'%ii, 'wb') as out:
        writer(out, parsed_schema, [record],codec='deflate')
```

```python
# main

import numpy as np
import matplotlib.pyplot as plt


cadance_1 = [0.,1/24*2.5,1/24*2.5*2]
cadence = np.array([(np.array(cadance_1)+i).tolist() for i in range(370)]).flatten()

maglim_survey = get_maglim(cadence,[-1.5])

# rates and luminosity change
# scale the lightcurves to typical mag abs 

kn_g, kn_r = model_kn()
grb_g, grb_r = model_grb()
SN1a_g, SN1a_r = model_sn1a()
TDE_g, TDE_r = model_tde()

models = {
    'kn':{
        'g_model' : kn_g,
        'r_model' : kn_r,
        'absmag' : -16.5
    },
    'grb':{
        'g_model' : grb_g,
        'r_model' : grb_r,
        'absmag' : -22 
    },
    'sn1a':{
        'g_model' : SN1a_g,
        'r_model' : SN1a_r,
        'absmag' : -18 
    },
    'tde':{
        'g_model' : TDE_g,
        'r_model' : TDE_r,
        'absmag' : -20 
    },
}

# transform absmag to distance limit to define volume visible
maglim_ZTF = 21

for t in models:
    distmod = maglim_ZTF - models[t]['absmag']  
    models[t]['distmod'] = distmod
    models[t]['distLim'] = 10**(distmod/5 + 1)/1e9
#     print(models[t]['distLim'] ,distmod)
    
# double check rates
rates = { # in /Gpc^3 /yr
    'tde' : 5,
    'kn' : 900,
    'grb' : 1,
    'sn1a' : 2.4e2
}

# get volume observed nightly ~ 1.5*pi^2 strad?
vol_night  = 1/3

# determine how many transients per category
for t in models:
    models[t]['N'] = rates[t] * (4/3*np.pi*models[t]['distLim'] ** 3) * vol_night
    print(t, models[t]['N'])
    
```

```python
# print(models['sn1a']['distLim'])

random.seed(10)

ii = 0

for t in models:

    print(t,int(models[t]['N']))
    plt.figure()
    for i in range(int(models[t]['N'])):
        # give a distance and transform to redshift
        ## can't be a uniform distribution - transfroming to spheric

        phi = random.uniform(0,2*np.pi)
        costheta = random.uniform(-1,1)
        v = random.uniform(0,1)
        theta = np.arccos( costheta )
        d_temp = models['sn1a']['distLim']*1e9 *  (v**(1/3))
        distmod_temp = 5*(np.log10(d_temp)-1)
        
        # 
        jdstart = int(random.uniform(0,370))
        sample_times = cadence[cadence>jdstart]-jdstart
        mag_lim_temp = maglim_survey[cadence>jdstart]
        mag_lim_temp = mag_lim_temp[sample_times<=120]
        sample_times = sample_times[sample_times<=120]

        filters = [[1,2,2] for i in range(int(len(sample_times)/3))]
        filters.append([1,2,2][:len(sample_times)%3])
        filters = np.array([item for sublist in filters for item in sublist])

        if not (len(filters)== len(sample_times)):
            print('check lenght of filters and cadence')

        r_jd = sample_times[np.where(filters==2)]
        g_jd = sample_times[np.where(filters==1)]

        #### modify filter cadence
        # light curve!
        # model - absmag + modulus@z

    #         if i==0: plt.plot(sample_times,models[t]['g_model'](sample_times))
        mag_all,jd_all = [],[]

        for f,jd in [('g_model',g_jd),('r_model',r_jd)]:
            mag_temp = models[t][f](jd)
            scale = np.min(mag_temp) - models[t]['absmag']
            Mag_temp = mag_temp-scale
            mag_temp = Mag_temp + distmod_temp

        #get times for two different bands
            mag_all.append(mag_temp.tolist())
            jd_all.append(jd.tolist())
            plt.plot(jd,mag_temp)
        
        jd_all = np.array([item for sublist in jd_all for item in sublist])  
        mag_all= np.array([item for sublist in mag_all for item in sublist])[np.argsort(jd_all)]        
        jd_all = jd_all + jdstart

        lc_temp = [jd_all,mag_all,filters,mag_lim_temp,t]
        
        if any(mag_all<mag_lim_temp):
            create_arvo(lc_temp,ii)
        else:
            print('not detected')
        ii = ii+1
    plt.gca().invert_yaxis()
    # transform lc to avro alerts using the maglim array
```
